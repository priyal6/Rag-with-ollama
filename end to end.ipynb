```python
import requests
from bs4 import BeautifulSoup

# Target website (replace with a real one)
url = "https://www.tpointtech.com/python-tutorial"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Extract all text from paragraphs
data = [p.text for p in soup.find_all('p')]
text_data = " ".join(data)

print("Extracted Data:", text_data[:1000])  # Print first 500 characters for verification
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100)
docs = splitter.split_text(text_data)

print(f"Total Chunks: {len(docs)}")
```

```python
from langchain.embeddings import HuggingFaceBgeEmbeddings
import chromadb

embedding_model = HuggingFaceBgeEmbeddings()

embeddings = [embedding_model.embed(doc) for doc in docs]

db = chromadb.PersistentClient(path="./chroma_db")

















```
